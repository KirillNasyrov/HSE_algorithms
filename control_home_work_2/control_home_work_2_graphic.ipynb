# Как я строил графики
Я сохранял тестирования производительности каждого алгоритма в файле csv. Формат такой:<br>
`Number of substitution;Size of text;Size of alphabet;Size of pattern;Type of algorithm;Time(ns)` <br>
Это мне позволило с помощью функций питона наиболее удобно отбирать нужные нам ячейки и строить графики. <br>
В папке с "Control_Homework_2_code" есть все реализованные алгоритмы, в main.cpp же происходит генерация текстов и шаблонов и тестирования алгоритмов с записью результатов в файл.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("fileOfTimes.csv", sep=";")

zero = data[(data["Number of substitution"] == 0) & (data["Size of text"] == 10000) & (data["Size of alphabet"] == 2)]

different_algorithms = zero['Type of algorithm'].unique()

for algorithm in different_algorithms:
    cur_algorithm = zero[zero['Type of algorithm'] == algorithm]

    plt.plot(cur_algorithm['Size of pattern'], cur_algorithm['Time(ns)'], label=algorithm)

plt.xlabel('Размер шаблона')
plt.ylabel('Время в наносек.')
plt.title('Без символов подстановки, 10000 символов, бинарный алфавит')

plt.legend()

plt.show()
```

## С символами подстановки

Я искал символы подстановки таким образом: для того чтобы найти все вхождения в текст заданного шаблона с масками Q, необходимо обнаружить вхождения в текст всех его безмасочных кусков.Пусть {Q1,…,Qk} — набор подстрок Q, разделенных масками, и пусть {l1,…,lk} — их стартовые позиции в Q. Для алгоритма понадобится массив C. C[i] — количество встретившихся в тексте безмасочных подстрок шаблона, который начинается в тексте на позиции i. Тогда появление подстроки Qi в тексте на позиции j будет означать возможное появление шаблона на позиции j−li+1.
   - Используя какой-то алгоритм нахождения стандартного шаблона в тексте, находим безмасочные подстроки шаблона Q: когда находим Qi в тексте T на позиции j, увеличиваем на единицу C[j−li+1].
   - Каждое i, для которого C[i]=k, является стартовой позицией появления шаблона Q в тексте. <br>

Рассмотрим подстроку текста T[i…i+n−1]. Равенство C[i]=k будет означать, что подстроки текста T[i+l1…i+l1+|Q1|−1],T[i+l2…i+l2+|Q2|−1] и так далее будут равны соответственно безмасочным подстрокам шаблона {Q1,…,Qk}. Остальная часть шаблона является масками, поэтому шаблон входит в текст на позиции i.

## стандартные алгоритмы
Практически на всех графиках видно, что навиный алгоритм работает эффективнее, чем два схожих алгоритма КМП. Есть несколько причин, почему это так.<br>
   - Наивный алгоритм выполняет меньше операций и требует меньше памяти для хранения данных, чем более сложные алгоритмы.
   - Наивный алгоритм может быть эффективнее на случайных или неструктурированных данных, так как он выполняет простой перебор всех возможных позиций в тексте, не зависящий от его структуры.
   - Применение стандартных или уточнённых граней в алгоритме Кнута-Морриса-Прата может занимать дополнительное время и память для их вычисления, что может привести к увеличению времени работы алгоритма.
   
Если посчитать мат ожидание того, сколько вхождений у нас будет при шаблоне размера 100 символов и текста размера 100000 символов, при бинарном алфавите, то оно будет равно (10^5 / 2^100), это число близко к 0. Учитывая, что мы взяли шаблон прямо из текста, то практически всегда по вероятности у нас будет только одно вхождение. А если шаблон и алфавит будет ещё больше, то мат ожидание будет ещё меньше. Так получается из-за того, что текст генерируется случайным образом. <br>
Наивный алгоритм проверяет каждое возможное вхождение шаблона в текст, поэтому его производительность зависит от количества вхождений шаблона в текст. Если сгенерированный текст случайным образом, то вероятность вхождения шаблона будет невысокой, и наивный алгоритм сможет быстро пройти через весь текст, не найдя вхождений. В то же время, алгоритм Кнута-Морриса-Прата с применением стандартных граней работает плохо, потому что в этом случае алгоритм должен выполнить полное сравнение шаблона со всеми подстроками текста. Таким образом, время выполнения алгоритма будет равно O(n * m), где n - длина текста, а m - длина шаблона. Аналогично и с алгоритмом Кнута-Морриса-Прата с применением уточненных граней <br>
С другой стороны, если вхождения шаблона в тексте есть, то алгоритм Кнута-Морриса-Прата с применением стандартных граней сможет использовать префикс-функцию для сокращения количества сравнений и выполнить поиск быстрее. Однако, если вхождения шаблона в тексте нет, то использование префикс-функции не помогает, и алгоритм снова приходится выполнять полное сравнение со всеми подстроками текста, что приводит к существенному увеличению времени выполнения.<br>
## доп алгоритм
Можно заметить, что на маленьком тексте автомат кмп работает медленно по сравнению с остальными алгоритмами, а на большом тексте наоборот, очень быстро. Это связано с тем, что на большом тексте стандартные алгоритмы кмп будут занимать гораздо больше дополнительной памяти. автомат также позволяет искать вхождения шаблона в тексте, используя только постоянный объем дополнительной памяти, что делает его достаточно эффективным для обработки больших объемов данных. Кроме того, автомат КМП может быть предварительно построен и сохранен, что позволяет многократно использовать его для поиска различных шаблонов в том же тексте, что может быть полезно при решении многих задач. Это мы наблюдается, когда мы ищем шаблон с символами подстановки. Однако, если текст небольшой и дан заранее, алгоритм Кнута-Морриса-Пратта будет работать быстрее.
