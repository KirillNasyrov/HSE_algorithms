# Вершины

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("results.csv", sep=",")
complete_data = data[(data["Graph"] == "Complete")]

vertex = complete_data["NumVertices"]

TimeDijkstra = complete_data["TimeDijkstra"]
TimeFloydWarshall = complete_data["TimeFloydWarshall"]
TimeBellmanFord = complete_data["TimeBellmanFord"]
TimeAStar = complete_data["TimeAStar"]


plt.plot(vertex, TimeDijkstra, label='Dijkstra')
plt.plot(vertex, TimeFloydWarshall, label='FloydWarshall')
plt.plot(vertex, TimeBellmanFord, label='BellmanFord')
plt.plot(vertex, TimeAStar, label='AStar')

plt.xlabel('Число вершин')
plt.ylabel('Время в наносек.')
plt.title('Полные графы')

plt.legend()

plt.show()
```
