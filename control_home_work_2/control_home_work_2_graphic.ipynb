# Без символа подстановки

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("fileOfTimes.csv", sep=";")

zero = data[(data["Number of substitution"] == 0) & (data["Size of text"] == 10000) & (data["Size of alphabet"] == 2)]

different_algorithms = zero['Type of algorithm'].unique()

for algorithm in different_algorithms:
    cur_algorithm = zero[zero['Type of algorithm'] == algorithm]

    plt.plot(cur_algorithm['Size of pattern'], cur_algorithm['Time(ns)'], label=algorithm)

plt.xlabel('Размер шаблона')
plt.ylabel('Время в наносек.')
plt.title('Без символов подстановки, 10000 символов, бинарный алфавит')

plt.legend()

plt.show()
```

# Вывод

Тестировал алгоритмы в конфигурации сборки Release. Это связано с тем, что в этом режиме код программы оптимизирован для максимальной производительности, что позволяет уменьшить время выполнения. <br>
Практически на всех графиках видно, что навиный алгоритм работает эффективнее, чем два схожих алгоритма КМП. Есть несколько причин, почему это так.<br>
   - Наивный алгоритм выполняет меньше операций и требует меньше памяти для хранения данных, чем более сложные алгоритмы.
   - Наивный алгоритм может быть эффективнее на случайных или неструктурированных данных, так как он выполняет простой перебор всех возможных позиций в тексте, не зависящий от его структуры.
   - Применение стандартных или уточнённых граней в алгоритме Кнута-Морриса-Прата может занимать дополнительное время и память для их вычисления, что может привести к увеличению времени работы алгоритма.
   
Вероятность вхождения шаблона в текст зависит от длины текста, длины шаблона и используемого алфавита. Если алфавит содержит много символов, то вероятность вхождения шаблона в текст будет ниже, чем если алфавит содержит меньше символов. <br>
Таким образом, если текст был сгенерирован случайным образом, то вероятность вхождения шаблона будет примерно одинаковой для всех позиций текста. Если шаблон был довольно коротким, то наивный алгоритм мог бы работать быстрее, чем алгоритм Кнута-Морриса-Прата, потому что наивный алгоритм выполняет не более, чем n * k операций для текста длиной n и шаблона длиной k, в то время как алгоритм Кнута-Морриса-Прата требует предварительной обработки шаблона и может потребовать больше времени на эту обработку. Однако, если шаблон был длинным и содержал повторяющиеся участки, то алгоритм Кнута-Морриса-Прата мог бы работать быстрее, потому что он может избежать повторных проверок этих участков благодаря использованию префиксных функций.
